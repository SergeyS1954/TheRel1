График изменения значений параметра $\pi_N$ для $N = 2,3,4$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

ro = np.arange (0, 2.0, 0.05)
N = 2
pi_N_N2  = ro**N/(1 + ro)**N
N = 3
pi_N_N3  = ro**N/(1 + ro)**N
N = 4
pi_N_N4  = ro**N/(1 + ro)**N
fig = plt.figure()   # Отрисовка данных
plt.plot (ro,pi_N_N2, linestyle=":", label = 'pi_N_N2')
plt.plot (ro,pi_N_N3, linestyle="-.", label = 'pi_N_N3')
plt.plot (ro,pi_N_N4, label = 'pi_N_N4')
plt.title('Расчет стационарной вероятности  нахождения в ремонте всех элементов') 
plt.xlabel('Отношение интенсивностей')   
plt.ylabel('Вероятность неготовности системы') 
plt.legend()  
plt.grid(True)
plt.show()
```
