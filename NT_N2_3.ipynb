График изменения значений среднего числа элементов, находящихся на обслуживании (в ремонте) $ N_T$ для $N = 2,3$ 

```python
import numpy as np
import matplotlib.pyplot as plt
import math

ro = np.arange (0, 2.0, 0.05)
N2_T  = (2*ro*(1 + 2*ro))/(1 + 2*ro + 2*ro**2)

N3_T = (3*ro*(1 + 4*ro + 6*ro**2))/(1 + 3*ro + 6*ro**2 + 6*ro**3)
fig = plt.figure()   # Отрисовка данных
plt.plot (ro,N2_T, '-', label = 'N2_T')
plt.plot (ro,N3_T, ':', label = 'N3_T')
plt.title('Расчет среднего число элементов в ремонте')  
plt.xlabel('Отношение интенсивностей')  
plt.ylabel('Среднее число')   
plt.grid(True)
plt.legend() 
plt.show()
```
