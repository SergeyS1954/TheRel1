Графики вероятностей неготовности системы для $M = 1,2,3$.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
t0  = 1.0
k = 0
ro = np.arange (0, 2.0, 0.05)
pi_0_M1  = 1/(1 + 3*ro + 6*ro**2 + 6*ro**3)
pi_3_M1 = pi_0_M1*6*ro**3
pi_0_M2  = 1/(1 + 3*ro + 3*ro**2 + 3/2*ro**3)
pi_3_M2 = pi_0_M2*(3/2)*ro**3
pi_0_M3  = 1/(1 + 3*ro + 3*ro**2 + ro**3)
pi_3_M3 = pi_0_M3*ro**3
fig = plt.figure()   # Отрисовка данных
plt.plot (ro,pi_3_M1, linestyle=":", label = 'pi_3_M1')
plt.plot (ro,pi_3_M2, linestyle="-.", label = 'pi_3_M2')
plt.plot (ro,pi_3_M3, label = 'pi_3_M3')
plt.title('Расчет стационарной вероятности неготовности системы') 
plt.xlabel('Отношение интенсивностей')   
plt.ylabel('Вероятность неготовности') 
plt.legend()  
plt.grid(True)
plt.show()
```
